```python
# Function to create time sequences for LSTM
def create_sequences(X, y, steps):
    Xs, ys = [], []
    for i in range(len(X) - steps):
        Xs.append(X[i:i+steps])
        ys.append(y[i+steps])
    return np.array(Xs), np.array(ys)
```

```python
# Sort by company and date for lstm
df_lstm = df.sort_values(['company', 'published_date']).reset_index(drop=True)
# Convert date to ordinal
df_lstm['date_ordinal'] = df_lstm['published_date'].apply(lambda x: x.toordinal())
```

```python
# Scale data between 0 and 1
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
scaled_features = scaler_X.fit_transform(df_lstm[features])
scaled_target = scaler_y.fit_transform(df_lstm[[target]])
```

```python
#Creating lists for hyperparameter tuning of the lstm model
steps_list = [30, 60, 90]
units_list = [50, 100]
dropout_list = [0.2, 0.3]
epochs_list = [5,10]
```

```python
#Creating a function for building the lstm model
def build_lstm_model(units, dropout, steps, feature_len):
    model = Sequential([
        Input(shape=(steps, feature_len)),
        LSTM(units, activation='relu', return_sequences=True),
        Dropout(dropout),
        LSTM(units, activation='relu'),
        Dropout(dropout),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
results_list = []
for steps in steps_list:
    # Create sequences for this step size
    X_seq, y_seq = create_sequences(scaled_features, scaled_target, steps)
    split = int(len(X_seq) * 0.8)
    X_train_seq, X_test_seq = X_seq[:split], X_seq[split:]
    y_train_seq, y_test_seq = y_seq[:split], y_seq[split:]
    
    for units in units_list:
        for dropout in dropout_list:
            for epochs in epochs_list:
                print(f"Training LSTM: steps={steps}, units={units}, dropout={dropout}, epochs={epochs}")
                
                model = build_lstm_model(units, dropout, steps, len(features))
                
                # Train model
                model.fit(X_train_seq, y_train_seq, epochs=epochs, batch_size=32, verbose=1)
                
                # Predict on test set
                pred_scaled = model.predict(X_test_seq)
                pred = scaler_y.inverse_transform(pred_scaled)
                actual = scaler_y.inverse_transform(y_test_seq.reshape(-1, 1))
                
                # Calculate metrics
                mse = mean_squared_error(actual, pred)
                rmse = np.sqrt(mse)
                mae = mean_absolute_error(actual, pred)
                r2 = r2_score(actual, pred)
                
                # Store results
                results_list.append([steps, units, dropout, epochs, mse, rmse, mae, r2])
```

```python
#viewing the results of all the lstm models 
results_df = pd.DataFrame(
    results_list,
    columns=['steps','units','dropout','epochs','MSE','RMSE','MAE','R2']
)
results_df
```

```python
#checking the best lstm model according to Mean squared error
best_mse_row = results_df.loc[results_df['MSE'].idxmin()]
best_mse_row
```

```python
#Checking the best lstm model according to Mean absolute error
best_mae_row = results_df.loc[results_df['MAE'].idxmin()]
best_mae_row
```

```python
#Checking the best lstm model according to the R2 score
best_r2_row = results_df.loc[results_df['R2'].idxmax()]
best_r2_row
```
