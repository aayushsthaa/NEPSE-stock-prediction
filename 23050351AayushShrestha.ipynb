LSTM

```python
# Function to create time sequences for LSTM
def create_sequences(X, y, steps):
    Xs, ys = [], []
    for i in range(len(X) - steps):
        Xs.append(X[i:i+steps])
        ys.append(y[i+steps])
    return np.array(Xs), np.array(ys)
```

```python
# Sort by company and date for lstm
df_lstm = df.sort_values(['company', 'published_date']).reset_index(drop=True)
# Convert date to ordinal
df_lstm['date_ordinal'] = df_lstm['published_date'].apply(lambda x: x.toordinal())
```

```python
# Scale data between 0 and 1
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
scaled_features = scaler_X.fit_transform(df_lstm[features])
scaled_target = scaler_y.fit_transform(df_lstm[[target]])
```

```python
# The model looks at the previous 30 days of data
# (open, high, low, per_change, traded_quantity, traded_amount, status, date)
# and uses that information to predict the closing price of the 31st day
steps = 30
# Create LSTM input-output sequences using past 'steps' days of data
# X_seq will store sequences of feature values
# y_seq will store the corresponding future close price
X_seq, y_seq = create_sequences(scaled_features, scaled_target, steps)
```

```python
# Calculate index where 80% of the data ends
split = int(len(X_seq) * 0.8)
# Use first 80% of sequence as X_train to trian the data and the rest 20% for the X_test
X_train, X_test = X_seq[:split], X_seq[split:]
# Use first 80% of sequence as y_train to trian the data and the rest 20% for the y_test
y_train, y_test_scaled = y_seq[:split], y_seq[split:]
```

```python
# Create a Sequential model (layers are stacked one after another)
model = Sequential([
    # Define input shape for LSTM
    # steps = number of past time steps
    # len(features) = number of features per time step
    Input(shape=(steps, len(features))),
    # LSTM layer with 50 memory units
    # Learns patterns from the time-series data
    # activation='relu' introduces non-linearity
    LSTM(50, activation='relu'),
    # Dense layer with 1 neuron
    # Outputs the predicted close price
    Dense(1)
])
```

```python
# Compile the model by usigng the adam optimizer and loss 
# Adam optimizer adapts learning rate automatically
# Mean Squared Error measures prediction error
model.compile(optimizer='adam', loss='mse')
```

```python
# Train the LSTM model using training data
# Input sequences for training
# Actual target values
# Number of times the model sees the full dataset
# Number of samples processed at once
# Displays training progress, loss per epoch
model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)
```

```python
# Predict close prices for test data (scaled values)
pred_scaled = model.predict(X_test)
# Convert scaled predictions back to original close price values
pred = scaler_y.inverse_transform(pred_scaled)
# Convert scaled actual test values back to original values
actual = scaler_y.inverse_transform(y_test_scaled.reshape(-1, 1))
```

```python
# Calculate Root Mean Squared Error (measures large errors more heavily)
rmse_lstm1 = np.sqrt(mean_squared_error(actual, pred))
# Calculate Mean Absolute Error (average absolute difference)
mae_lstm1 = mean_absolute_error(actual, pred)
# Calculate R-squared score (how well predictions match actual values)
r2_lstm1 = r2_score(actual, pred)
```

```python
#print the metrics
print("LSTM Model Performance:")
print(f"RMSE: {rmse_lstm1:.4f}")
print(f"MAE:  {mae_lstm1:.4f}")
print(f"R2:   {r2_lstm1:.4f}")
```
